```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
%matplotlib inline
```

```python
# Load data
all_crime_borough = pd.read_table('results/all_offense_borough.out',sep='\t', \
                          header=None, names=['Borough','Count']) \
                          .sort_values(by='Count')

all_crime_day = pd.read_table('results/all_offense_hour_from.out', \
                                  delimiter='\t', names=['Hour','Count'])

all_crime_dow = pd.read_table('results/all_offense_day_of_week.out', \
                                  delimiter='\t', names=['Day of Week','Count'])

all_crime_month = pd.read_table('results/all_offense_month_from.out', \
                                  delimiter='\t', names=['Month','Count'])

all_crime_year = pd.read_table('results/all_offense_year.out', \
                                  delimiter='\t', names=['Year','Count'])

all_crime_year_last_decade = all_crime_year.loc[all_crime_year['Year']>=2006]
```

# PLOTS

## All Crime in Different Boroughs

```python
sns.set(style="white", context="talk")

# Set up the matplotlib figure
fig, ax = plt.subplots()

sns.barplot(all_crime_borough['Borough'], all_crime_borough['Count'], palette="YlOrRd")
ax.set_ylabel("All Crime by Boroughs")

# Finalize the plot
sns.despine(bottom=True, left=True)

#plt.setp(fig.axes, yticks=[]) # Remove y-labels
```

## All Crime throughout the Day

```python
sns.set(style="white", context="talk")

# Set up the matplotlib figure
fig, ax = plt.subplots()

sns.barplot(all_crime_day['Hour'][1:25], all_crime_day['Count'], palette="Blues_r")
ax.set_ylabel("Complaints")
plt.title('All Crime Complaints by Time of Day (2006-2015)')

# Finalize the plot
sns.despine(bottom=True, left=True)
#plt.setp(fig.axes, yticks=[]) # Remove y-labels
```

## All Crime throughout the Week

```python
sns.set(style="white", context="talk")

# Set up the matplotlib figure
fig, ax = plt.subplots()

day_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

sns.barplot(day_of_week, all_crime_dow['Count'][:7], palette="Blues_r")
ax.set_xlabel("Day of Week")
ax.set_ylabel("Complaints")
plt.ylim(500000,)
plt.title('All Crime Complaints Throughout the Week (2006-2015)')

# Finalize the plot
sns.despine(bottom=True, left=True)
```

## All Crime throughout the Year

```python
sns.set(style="white", context="talk")

# Set up the matplotlib figure
fig, ax = plt.subplots()

month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

sns.barplot(month, all_crime_month['Count'][1:], palette="Blues_r")
ax.set_xlabel("Month")
ax.set_ylabel("Complaints")
plt.title('All Crime Complaints by Month (2006-2015)')
plt.ylim(250000,475000)

# Finalize the plot
sns.despine(bottom=True, left=True)
```

## All Crime in the Last Decade

```python
sns.set(style="white", context="talk")

# Set up the matplotlib figure
fig, ax = plt.subplots()

sns.barplot(all_crime_year_last_decade['Year'], all_crime_year_last_decade['Count'], palette="Blues_r")
ax.set_ylabel('Complaints')
plt.ylim(450000,550000)
plt.title("All NYC Crime Complaints: 2006-2015")


# Finalize the plot
sns.despine(bottom=True, left=True)
```
