```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load key values
key_borough_crime = pd.read_table('results/key_code_boro.out', delimiter='\t', \
                                  names=['Code','Borough','Count']).dropna()
popular_crime = key_borough_crime.loc[key_borough_crime['Count']>10000]
really_popular_crime = popular_crime.loc[popular_crime['Count']>50000]
```

# Crime by Code in Different Boroughs

## All Codes

```python
sns.set(style="white", context="talk")
plt.figure(figsize=(20, 20))
sns.set_palette(sns.color_palette("muted", desat=.4))
sns.barplot(x="Code", y="Count", hue="Borough", data=key_borough_crime, \
              estimator = sum, palette=sns.color_palette("cubehelix", 7))

sns.despine(bottom=True, left=True)
```

## Crimes with More than 10000 Violations

```python
sns.set(style="white", context="talk")
sns.set_palette(sns.color_palette("muted", desat=.4))
sns.barplot(x="Code", y="Count", hue="Borough", data=popular_crime, \
              estimator = sum, palette=sns.color_palette("cubehelix", 7))
sns.despine(bottom=True, left=True)
```

## Crimes with More than 50000 Violations

```python
fig, ax = plt.subplots()
sns.set(style="white", context="talk")
sns.set_palette(sns.color_palette("muted", desat=.4))
sns.barplot(x="Code", y="Count", hue="Borough", data=really_popular_crime, \
              estimator = sum,palette=sns.color_palette("cubehelix", 7))
ax.set_ylabel('Complaints')
plt.title('Crime Complaints: Key Codes with more than 50,000 Complaints')
sns.despine(bottom=True, left=True)
```
